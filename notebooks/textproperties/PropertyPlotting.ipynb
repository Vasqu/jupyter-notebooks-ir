# Plotting the properties of texts

```python
%matplotlib inline
from matplotlib import pyplot as plt
import math
```

## Zipf's Law

Comparison of predictions (=$\frac{1}{n(n+1)}$) and actual results on 1GB TREC Volume 3 Corpus:

```python
total_unique_words = 502209
actual_proportions = [.402, .132, .069, .046, .032, .024, .019, .016, .014, .012]
predicted_proportions = [round((1/(n*(n+1))), 3) for n in range(1,11,1)]

x = [n for n in range(1,11,1)]
y = [math.ceil(p*total_unique_words) for p in actual_proportions]
z = [math.ceil(p*total_unique_words) for p in predicted_proportions]

plt.plot(x, y, label="actual_counts")
plt.plot(x, z, label="predicted_counts")
plt.legend(loc='upper right')
plt.ylabel('number of words')
plt.xlabel('number of occurrences')
plt.title('1GB TREC Volume 3 Corpus - Occurrence frequencies')
plt.show()
```

Plotting Zip's law $\rightarrow y = kx^{-1}$:

```python
x = [x for x in range(1,100,1)]
# here k=100
y = [100*math.pow(x,-1) for x in range(1,100,1)]

plt.plot(x, y)
plt.ylabel('frequency')
plt.xlabel('rank')
plt.title('Zipf\'s Law - Raw plot')
plt.show()

plt.loglog(x, y)
plt.ylabel('frequency')
plt.xlabel('rank')
plt.title('Zipf\'s Law - Log-log plot')
plt.show()
```

## Heap's Law

Plotting Heap's law $\rightarrow V = KN^{\beta}$:

```python
x = [x for x in range(1,100,1)]
# here K=50, beta=0.4
y = [50*math.pow(x, 0.4) for x in range(1,100,1)]

plt.plot(x, y)
plt.ylabel('vocabulary size')
plt.xlabel('words in collection')
plt.title('Heap\'s Law - Raw plot')
plt.show()

plt.loglog(x, y)
plt.ylabel('vocabulary size')
plt.xlabel('words in collection')
plt.title('Heap\'s Law - Log-log plot')
plt.show()
```
